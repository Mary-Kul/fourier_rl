```python
eval_env = NormalizedBoxEnv(HalfCheetahEnv())
```

```python
def get_action_mean_ft_spectrum_from_epoch(base_dir,act_num,epoch_num):
    print(os.path.join(base_dir,f'itr_{epoch_num}.pkl'))
#     torch.device('cpu') 
    agent = torch.load(os.path.join(base_dir,f'itr_{epoch_num}.pkl'),map_location='cpu' )
    agent = agent['trainer/policy']

    rollout_dict = rollout(
    env = eval_env ,
    agent = agent,
    max_path_length=1000,
    render=False,
    render_kwargs=None,
        )

#     M = int(10*8)
#     slices = util.view_as_windows(rollout_dict['actions'][:,act_num], window_shape=(M,), step=6)
#     # remove DC
#     slices -= np.mean(slices)
    
#     win = np.hanning(M + 1)[:-1]
#     slices = slices * win

#     spectrum = np.fft.rfft(slices.T, axis=0, norm="ortho")

    spectrum = torch.stft(torch.from_numpy(rollout_dict['actions'][:,act_num]),n_fft = 80,hop_length = 6, win_length=40,normalized = True)
    spectrum = np.abs(spectrum).mean(axis =1)[:,0]
#     return (np.abs(spectrum).mean(axis = 1))
#     print(type(spectrum.numpy()))
    return spectrum.numpy()
```

```python
def save_fourier_rollouts(y_dict,name,dir_name = 'rolloutsa_base'):
    # Store data (serialize)
    with open(f'dicts_w_rollouts/{dir_name}/{name}', 'wb') as handle:
        pickle.dump(y_dict, handle, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
y_dict ={}
max_steps = 2990
for j in range(9):
    for act in range(6):
        for epoch in range(0,max_steps,10):
            y_dict[f'{act}_{epoch}'] = get_action_mean_ft_spectrum_from_epoch(base_dir = base_dir,act_num = act,
                                                                              epoch_num = epoch)
        
    save_fourier_rollouts(y_dict,name = f'base_run_1-c34bcdf8_{j+1}.pkl')
```
